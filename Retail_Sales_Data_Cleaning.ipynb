```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv(r"C:\Users\susmi\Downloads\retail_sales_dataset.csv")
df
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df
```

```python
df['Month-Year'] = df['Date'].dt.to_period('M').astype(str)
df
```

```python
def age_group(age):
    if age < 18:
        return 'Under 18'
    elif 18 <= age <= 25:
        return '18-25'
    elif 26 <= age <= 35:
        return '26-35'
    elif 36 <= age <= 50:
        return '36-50'
    else:
        return '51+'
```

```python
df['Age Group'] = df['Age'].apply(age_group)
df
```

```python
df.dropna(inplace=True)
df
```

```python
df.to_csv(r"C:\Users\susmi\Downloads\Cleaned_Retail_Sales.csv", index=False)
```
